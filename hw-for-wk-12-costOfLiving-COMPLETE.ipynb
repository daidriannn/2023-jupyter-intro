**Note about social security increases:**

Each year the U.S. government increases the amount of social security retired people receive. That increase is based on the annual inflation rate  and the political climate (or party in control). I have indexed the numbers so 1995 is 100, and each year after that the number reflects a percentage increase since 1995. So between 1995 at 100 and 1996 at 102.6 reflects a 2.6 percent increase. And between 1995 and 1997 reflects a 5.6 percent increase.




```python
import pandas as pd
```

```python
df_raw = pd.read_csv("cable-bill-only - Sheet1.csv")
df_raw
```

```python
df_raw.info()
```

```python
df_raw.columns
```

```python
df = df_raw[['Avg. cable bill', 'Median household income','Social Security cost-of-living adjustment']].copy()
df
```

```python
df.drop(0).reset_index(drop = True)
```

```python
df.info()
```

```python
df["Median household income"] = pd.to_numeric(df["Median household income"], errors = "coerce")
df.info()
```

```python
df["Social Security cost-of-living adjustment"] = pd.to_numeric(df["Social Security cost-of-living adjustment"], errors = "coerce")
df.info()
```

```python
df["Avg. cable bill"] = pd.to_numeric(df["Avg. cable bill"], errors = "coerce")
df.info()
```

```python
df
```

```python
df.to_csv("cable-bill-only - Sheet1.csv" , index = False, encoding = "UTF8")
```

```python
avg_cable = df["Avg. cable bill"].iat[1]
avg_cable
```

```python
df["cable_pct_chg"] = \
((df["Avg. cable bill"] - avg_cable) / avg_cable) * 100

df
```

```python
med_income = df["Median household income"].iat[1]
med_income
```

```python
df["medincome_pct_chg"] = \
((df["Median household income"] - med_income) / med_income) * 100

df
```

```python
ss_cola = df["Social Security cost-of-living adjustment"].iat[1]
ss_cola
```

```python
df["sscola_pct_chg"] = \
((df["Social Security cost-of-living adjustment"] - avg_cable) / avg_cable) * 100

df
```

```python
df.plot(x= "Avg. cable bill",
       y = ["cable_pct_chg", "medincome_pct_chg","sscola_pct_chg"])
```

```python
df.plot(x = "Avg. cable bill",
       y = ["cable_pct_chg", "medincome_pct_chg","sscola_pct_chg"],
       title = "Trend: Cable; Median Income; Social Security Increases",
        style = {"cable_pct_chg": "red",
                 "medincome_pct_chg": "blue",
                 "sscola_pct_chg":"gray"})
```
